```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
output_notebook()
```

The home price data used is from Robert J. Shiller, PhD of Yale University: http://www.econ.yale.edu/~shiller/data.htm

```python
df = pd.read_excel('shiller_home_data.xlsx')
```

```python
df.head()
```

```python
p = figure(title="Real Home Price Index & Real Building Cost Index",
          x_axis_label='x', y_axis_label='y')
```

```python
p.line(df['Date'], df['Real Home Price Index'], legend_label="Price", line_width=2)
p.line(df['Date'], df['Real Building Cost Index'], legend_label="Cost", line_width=2)
p.line(df['Date'], df['US Population (Millions)'], legend_label="Population", line_width=2)
show(p)
```
