```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
output_notebook()
```

Pulling Crime data for Ohio from 1960-2019: https://www.disastercenter.com/crime/ohcrime.htm

```python
df = pd.read_excel('Ohio_Crime_Data.xlsx')
```

```python
df.tail()
```

```python
p = figure(title="Violent Crimes in Ohio",
          x_axis_label='Year', y_axis_label='Incidents',
          plot_width=900, plot_height=500)
```

```python
p.line(df['Year'], df['Violent'], legend_label="Violent Crime", line_width=2, color='red')
p.line(df['Year'], df['Murder'], legend_label="Murder", line_width=2, color='orange')
p.line(df['Year'], df['Aggrevated_assault'], legend_label="Assault", line_width=2, color='magenta')
show(p)
```
